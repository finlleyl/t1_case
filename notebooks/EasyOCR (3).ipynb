```python
import os
import shutil
import random
import pandas as pd
import csv

train_dir     = '/content/train'
labels_tsv    = '/content/train.tsv'

train_out_dir = '/content/EasyOCR/trainer/all_data/ru_train_filtered'
val_out_dir   = '/content/EasyOCR/trainer/all_data/ru_val'

split_ratio   = 0.8
random_seed   = 42


os.makedirs(train_out_dir, exist_ok=True)
os.makedirs(val_out_dir,   exist_ok=True)


with open(labels_tsv, 'r', encoding='utf-8') as f:
    lines = f.readlines()
random.seed(random_seed)
random.shuffle(lines)

split_idx   = int(len(lines) * split_ratio)
train_lines = lines[:split_idx]
val_lines   = lines[split_idx:]


def move_images(lines, src_dir, dst_dir):
    for line in lines:
        img_name = line.split('\t')[0].strip()
        src_path = os.path.join(src_dir, img_name)
        dst_path = os.path.join(dst_dir, img_name)
        if os.path.exists(src_path):
            shutil.move(src_path, dst_path)
        else:
            print(f'Файл не найден: {src_path}')

move_images(train_lines, train_dir, train_out_dir)
move_images(val_lines,   train_dir, val_out_dir)

def write_labels_csv(lines, target_dir, csv_name='labels.csv', header=True):
    rows = [ln.rstrip('\n').split('\t') for ln in lines]
    df   = pd.DataFrame(rows, columns=['filename', 'words'] if header else None)  # изменено название столбцов
    df.to_csv(os.path.join(target_dir, csv_name),
              index=False, header=header, encoding='utf-8', quoting=csv.QUOTE_MINIMAL)

write_labels_csv(train_lines, train_out_dir)
write_labels_csv(val_lines,   val_out_dir)

print(f'✓ Готово!\n  train  → {train_out_dir} ({len(train_lines)} файлов)\n'
      f'  val    → {val_out_dir}   ({len(val_lines)} файлов)')
```

```python

import os
import torch.backends.cudnn as cudnn
import yaml
from trainer.train import train
from utils import AttrDict
import pandas as pd

cudnn.benchmark = True
cudnn.deterministic = False

def get_config(file_path):
    with open(file_path, 'r', encoding="utf8") as stream:
        opt = yaml.safe_load(stream)
    opt = AttrDict(opt)
    if opt.lang_char == 'None':
        characters = ''
        for data in opt['select_data'].split('-'):
            csv_path = os.path.join(opt['train_data'], data, 'labels.csv')
            df = pd.read_csv(csv_path, sep='^([^,]+),', engine='python', usecols=['filename', 'words'], keep_default_na=False)
            all_char = ''.join(df['words'])
            characters += ''.join(set(all_char))
        characters = sorted(set(characters))
        opt.character= ''.join(characters)
    else:
        opt.character = opt.number + opt.symbol + opt.lang_char
    os.makedirs(f'./saved_models/{opt.experiment_name}', exist_ok=True)
    return opt

#Запускаем обучение
opt = get_config("config_files/custom_data_train.yaml")
train(opt, amp=False)
```
